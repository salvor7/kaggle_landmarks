```python
from source import data
import csv
import os
import pathlib
import time
```

Loop through the csv file which contains the list of all the image file names and open, resize, and save them to another folder

```python
#initialize variables
src_path = '/media/rsuprun/Alexandria3000/Kaggle/data/recognition/train_images/'
dst_path = data.path_string('train_images','448','') # I suggest a separate folder for each differently sized set
csv_file_path = data.path_string('recognition', 'train.csv')
trgt_size = 448

#specify the start and stop lines from the csv
startnum = 0
endnum = 1000

#result counters
cnt_total = 0
cnt_success = 0
cnt_skip = 0

tstart = time.time()
    
with open(csv_file_path) as training_set:
    reader = csv.reader(training_set)
    try: 
        for image, url in reader:
            if reader.line_num >= startnum and reader.line_num <= endnum:
                cnt_total += 1
                if os.path.exists(dst_path + image + '.jpg'): #skip the file if it already exists
                    cnt_skip += 1
                else:
                    cnt_success = data.resize_image(image + '.jpg', src_path, dst_path, trgt_size) + cnt_success
    
    except ValueError: #this is raised if there are too few arguments for the csv.reader, 
                        #which is the case if this the test.csv file since it has no landmark field
        for image, url, landmark in reader:
            if reader.line_num >= startnum and reader.line_num <= endnum:
                cnt_total += 1
                if os.path.exists(dst_path + image + '.jpg'): #skip the file if it already exists
                    cnt_skip += 1
                else:
                    cnt_success = data.resize_image(image + '.jpg', src_path, dst_path, trgt_size) + cnt_success
    
    except Exception as err:
        print(f'Warning: Failed because {err}')

telap = time.time() - tstart
print(f'{telap} seconds elapsed')
print(f'{cnt_success} of {cnt_total} images resized successfully.')
print(f'{cnt_skip} of {cnt_total} images skipped.')
    
```
