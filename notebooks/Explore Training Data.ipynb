```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt

% matplotlib inline
```

```python
data.path_string('recognition', 'train.csv')
```

```python
df_train = pd.read_csv(data.path_string('recognition', 'train.csv'))
```

```python
landmark_counts = df_train.groupby('landmark_id').count()
landmark_counts.head()
```

```python
distribution = landmark_counts.groupby('url').count().reset_index()
distribution.columns = ['samples', 'count']
distribution.tail()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)

sns.barplot(data=distribution, x='samples', y='count', ax=ax)

ax.set_xticklabels(labels=[]);
```

```python
distribution['logcount'] = np.log(distribution['count'])

fig, ax = plt.subplots()
fig.set_size_inches(20, 10)

sns.barplot(data=distribution, x='samples', y='logcount', ax=ax)

ax.set_xticklabels(labels=[]);
```
